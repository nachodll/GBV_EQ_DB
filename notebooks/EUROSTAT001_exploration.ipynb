# 1. Import Microdata from European Survey

```python
import os

import pandas as pd
```

```python
dataset_dir = os.path.join("data", "raw", "EUROSTAT", "EUROSTAT001-EncuestaEuropeaViolenciaGénero")
dataset_path = os.path.join(dataset_dir, "encuesta_europea_GBV.csv")
df = pd.read_csv(dataset_path, delimiter=";")
df
```

# 3. Extract codes and values

```python
vars = pd.DataFrame(columns=["Code", "UniqueValues"])
vars["Code"] = list(df.columns)
vars["UniqueValues"] = vars["Code"].apply(lambda cod: df[cod].unique())
vars
```

```python
vars.to_excel("vars.xlsx", index=False)
```
