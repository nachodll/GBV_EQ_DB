```python
import os

import pandas as pd
from dotenv import load_dotenv
```

```python
load_dotenv()
root_path = os.getenv("ROOT_PATH")
datasets_dir = os.path.join(root_path, "datasets", "preliminary_merge")

# List all CSV files in the directory
csv_files = [f for f in os.listdir(datasets_dir) if f.endswith(".csv")]

# Import each CSV as a pandas DataFrame and store in a dictionary
dataframes = {}
for file in csv_files:
    df_name = os.path.splitext(file)[0]
    dataframes[df_name] = pd.read_csv(os.path.join(datasets_dir, file))
```
